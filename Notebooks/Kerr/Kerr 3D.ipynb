```python
import numpy as np
from utils import Timelike, Nulllike, StaticGeodesicPlotter
```

```python
# Constant Radius Orbit
position = [4.0, np.pi/3, 0.0]
momentum = [0., 0.767851, 2.0]
a = 0.99
steps = 400
delta = 1.0
```

```python
geod = Timelike(
    metric="Kerr",
    metric_params=(a,),
    position=position,
    momentum=momentum,
    steps=steps,
    delta=delta,
    return_cartesian=True
)
```

```python
sgpl = StaticGeodesicPlotter()
sgpl.animate(geod, interval=100)
sgpl.show()
```
