```python
import numpy as np
from utils import Timelike, Nulllike, StaticGeodesicPlotter
```

```python
# Constant Radius Orbit
position = [3.5, np.pi/3, 0.]
momentum = [0., 0.767851, 2.0]
a = 0.0
steps = 100
delta = 0.05
```

```python
geod = Timelike(
    metric="Schwarzschild",
    metric_params=(a,),
    position=position,
    momentum=momentum,
    steps=steps,
    delta=delta,
    return_cartesian=True
)
```

```python
%matplotlib inline

sgpl = StaticGeodesicPlotter()
sgpl.animate(geod, interval=50)
sgpl.show()
```
