```python
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u
from einsteinpy.rays import Shadow
from einsteinpy.plotting import ShadowPlotter
```

```python
# Masas demasiado grandes dan algunos errores num√©ricos
mass = 1 * u.kg
fov = 20 * u.km
```

```python
shadow = Shadow(mass=mass, fov=fov, n_rays=5000)
```

```python
theta1 = np.linspace(0, 2 * np.pi, len(shadow.fb1))
r1, theta1 = np.meshgrid(shadow.fb1, theta1)
values1, values2 = np.meshgrid(shadow.intensity, shadow.intensity)

xs = r1 * np.cos(theta1)
ys = r1 * np.sin(theta1)

plt.figure(figsize=(10, 10))
plt.pcolormesh(xs, ys, values1, cmap=plt.cm.afmhot, shading="gouraud")
plt.axis('off')
plt.title('Schwarzschild black hole', weight='bold')

plt.show()
```
